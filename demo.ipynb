```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import plotstyle

from plotstyle.export import save_timestamped_figure
from plotstyle.colors import cambridge_light, cambridge_core, cambridge_dark
from plotstyle.typesetting import latex_is_available
from plotstyle.size import convert_unit, get_dim, set_dim, WIDTH

print(f"Latex is available: {latex_is_available()}")
```

Styles can be set via:

```python
with plt.style.context("nature"):
    fig, ax = plt.subplots(nrows=1, ncols=1, figsize=get_dim(WIDTH.nature_column))
    ax.plot([1, 2, 3], [1, 2, 3])
    ax.set_xlabel("x label")
    ax.set_ylabel("y label $\sum_i^n \cos(x)$")
    ax.set_title("Title")
    sns.despine()
    plt.show()
```

Styles can be combined with other styles or even dictionaries. Later elements in the list supersede earlier ones:

```python
with plt.style.context(["no-latex", "nature"]):
    fig, ax = plt.subplots(nrows=1, ncols=1, figsize=get_dim(WIDTH.nature_column, fraction_of_line_width=1.0))
    ax.plot([1, 2, 3], [1, 2, 3])
    ax.set_xlabel("x label")
    ax.set_ylabel("y label $\sum_i^n \cos(x)$")
    ax.set_title("Title")
    sns.despine()
    plt.show()
```

```python
with plt.style.context(["powerpoint", "pdflatex"]):
    fig, ax = plt.subplots(nrows=1, ncols=1, figsize=get_dim(WIDTH.powerpoint_standard, fraction_of_line_width=0.3))
    ax.plot([1, 2, 3], [1, 2, 3])
    ax.set_xlabel("x label")
    ax.set_ylabel("y label $\sum_i^n \cos(x)$")
    ax.set_title("Title")
    plt.show()
```
